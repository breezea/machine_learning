```python
# df = pd.read_csv('./reuters.csv',header=0, usecols=[0,1,2,3,4,5])
df = pd.read_csv('../data/reuters.csv',header=0,)
df = df.dropna(0,how='all', subset=['TOPICS'] )
df.head()
```

```python
topics = df.iloc[:, ]
topics = topics.value_counts()
topics = dict(topics)
# topics
```

```python

store = {}
# count = 0
def set_store(topic, title, body):
    if store.get(topic,False):
        store[topic][title] = body
    else:
        store[topic] = {}
        store[topic][title] = body
        
for row in df.itertuples():
    # print(row)
    index = row[0]
    topics = row[3]
    title = row[5]
    body_row = row[6:]
    body = ''
    for s in body_row:
        if pd.isnull(s):
            break
        else:
            body += s
    for topic in topics.split(','):
        if topic:
            set_store(topic, title, body)
    # if count > 8:
    #     break
    # count += 1
# print(store.keys())
# print(store['cocoa'].keys())
# print(store['cocoa,'].keys())
# print(store['cocoa,'].values())
```

```python
topics = store.keys()
counter = {}
topic_counter = {}
predict_counter = {}
feature_num = 20

def set_counter(topic, word):
    if not word: return
    if counter.get(topic, False):
        if counter[topic].get(word, False):
            counter[topic][word] += 1
        else:
            counter[topic][word] = 1
        counter[topic]['VAR_TOTAL'] += 1
    else:
        counter[topic] = {'VAR_TOTAL': 1}
        counter[topic][word] = 1

def set_topic_counter(topic):
    if topic_counter.get(topic, False):
        topic_counter[topic] += 1
        topic_counter['VAR_TOTAL'] += 1
    else:
        topic_counter[topic] = 1
        topic_counter['VAR_TOTAL'] = 1

# topics
for topic in store.keys():
    set_topic_counter(topic)
    for title,body in store[topic].items():
        words = body.split(' ')
        for word in words:
            set_counter(topic, word)

for topic in store.keys():
    topic_info = list(counter[topic].items())
    topic_info.sort(key = lambda x: x[1], reverse=True)
    feature_words = dict(topic_info[1: feature_num])
    feature_words['VAR_TOTAL'] = sum(feature_words.values())
    predict_counter[topic] = feature_words
    # predict_counter[topic] = dict(topic_info[1:feature_num])
    # print(topic, topic_info)
    # print(topic_info[1:feature_num], dict(topic_info[1:feature_num]))
    # break

def predict(sentence,n=3):
    res = []
    words = sentence.split(' ')
    for topic in topics:
        topic_p= topic_counter[topic] / topic_counter['VAR_TOTAL']
        score = 1
        for word in words:
            topic_word_p = predict_counter[topic].get(word,0) / predict_counter[topic]['VAR_TOTAL']
            score *= topic_word_p
        res.append({
            'topic': topic,
            'score': topic_p * score
        })
    res.sort(key=lambda x: x['score'], reverse=True)
    return res[:n]

# print(counter)
# print(counter.keys())
# print(topic_counter)
# predict('Showers continued throughout')
predict('and')
# print(counter['cocoa,']['continued'])
```
